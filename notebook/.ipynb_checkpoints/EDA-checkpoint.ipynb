```python
class KMeansCluster2D:
    def __init__(self, K, points):
        self.K = K
        self.points = points
        M, N = points.shape
        assert N == 2
        self.centeroids = np.zeros((K, N))
        self.classifications = np.zeros((M,))
    
    
    def assignPoints(self):
        M, N = self.points.shape
        classifications = np.zeros((M,))
        
        for i in range(M):
            smallestDistance = np.inf
            point = self.points[i]
            for k in range(self.K):
                centeroid = self.centeroids[k]
                distance = np.linalg.norm(point-centeroid)
                if distance < smallestDistance:
                    classifications[i] = k # assign point i to cluster k
                    smallestDistance = distance
        
        return classifications
    
    
    def recalcCentroids(self):
        M, N = points.shape
        newCenteroids = np.zeros((self.K, N))
        for k in range(self.K):
            indices = (self.classifications == k)
            total_points = sum(indices) # total points belonging to cluster k
            if total_points > 0: # if there is at least one point assigned to cluster k
                # calculate the mean of centeroid k and set this as the new centeroid
                newCenteroids[k,:] = np.sum(self.points[indices,:], axis=0) / total_points
            else: # if no point are assigned to cluster k
                # cluster k remains unchanged
                newCenteroids[k,:] = self.centeroids[k,:]
        return newCenteroids
```

```python
points = np.hstack((w.reshape(-1, 1),h.reshape(-1, 1)))
points.shape
```

```python
fig = plt.figure(figsize=(15,8))
km = KMeansCluster2D(3, points)
previous_centeroids = km.centeroids - 1
iteration = 0
while not (km.centeroids == previous_centeroids).all():
    iteration += 1
    previous_centeroids = km.centeroids
    km.classifications = km.assignPoints()
    km.centeroids = km.recalcCentroids()
    plt.cla()
    plt.title(f'{iteration}')
    plt.plot(w, h, '.k')
    plt.plot(km.centeroids[:,0], km.centeroids[:,1], 'or', markersize=10)
    display(fig)
    clear_output(wait = True)

# print the anchor boxes
# sort the centeroids
ind = np.lexsort((km.centeroids[:,1], km.centeroids[:,0]))
km.centeroids = km.centeroids[ind]
print("Anchor Boxes")
print(km.centeroids)
```

## Computing $t_x$ and $t_y$

The equation for the $x$ and $y$ coordinate of the bounding box is given below:

$$
b_x = \sigma{(t_x)} + c_x \\
b_y = \sigma(t_y) + c_y
$$

The above equation represents the $x$ and $y$ center of out ground truth bounding box denoted by the variables $b_x$ and $b_y$. The job of our model in predicting the location of the bounding box is predicting the center of the box. In the above equation, the prediction of our model is $t_x$ and $t_y$. We need to first calculate $t_x$ and $t_y$ from our dataset using the following equations. In the following equations we assume $(c_x, c_y) \approx (0, 0)$

$$
t_x = \sigma^{-1}(b_x) \\
t_y = \sigma^{-1}(b_y)
$$

In both sets of equations $c_x$ and $c_y$ are the location of the grid cell predicting the bounding box. For example, if the gird cell in the $7^{th}$ row and $7^{th}$ column is predicting the bounding box, then $(c_x, c_y)=(7,7)$. The inverse of the sigmoid function is defined below:

$$
\sigma^{-1}(x) = \ln\bigg(\frac{x}{1-x}\bigg)
$$

In the above equations, the range of $b_x\in[0,1]$ and $b_y\in[0,1]$

One thing to note is that in the equation for calculating $b_x$ and $b_y$, due to the addition of $c_x$ and $c_y$, which are greater than 1, the result will be greater than 1. Thus we need to normalize $b_x$ and $b_y$ by dividing by the number of rows in the grid. We can use either the number of rows or the number of columns, since the grids must always be squares.

```python
inverse_sigmoid = np.vectorize(lambda x: np.log(x/(1-x)))
```

```python
csv = pd.read_csv('augmented_train.csv')
csv = csv[csv['x_min'].notna()]

def get_tx_ty_tw_th(filename):
    bbox_coord = csv[csv['image_id']==filename].values[:, 4:]  
    
    x = bbox_coord[:, ::2]
    y = bbox_coord[:, ::2]
    
    w = x[:,1] - x[:,0]
    h = y[:,1] - y[:,0]
    
    txs = inverse_sigmoid((x[:,1]+x[:,0])/2)
    tys = inverse_sigmoid((y[:,1]+y[:,0])/2)
    
    for i, (pw, ph) in enumerate(anchor_boxes):
        tws = np.log((w/pw).astype(float))
        ths = np.log((h/ph).astype(float))
        

    return txs, tys
```

```python
def generate_target(y_train, anchor_boxes, anchor_mask, grid_size=12):
    box_wh = y_train[:, 2:4] - y_train[:, 0:2]

    # compute the intersection of each box with each anchor
    bw = box_wh[:,0].reshape(-1, 1)
    iw = np.minimum(np.tile(bw, [1, anchor_boxes.shape[0]]), anchor_boxes[:,0])
    bh = box_wh[:,1].reshape(-1, 1)
    ih = np.minimum(np.tile(bh, [1, anchor_boxes.shape[0]]), anchor_boxes[:,1])
    intersection = iw * ih

    # Compute the union of each box with each anchor
    box_area = box_wh[:,0] * box_wh[:,1]
    box_area = np.tile(box_area.reshape(-1, 1), [1, 12])
    anchor_area = anchor_boxes[:,0] * anchor_boxes[:,1]
    union = box_area + anchor_area - intersection

    # compute the iou of each box with each anchor
    iou = intersection / union

    #
    anchor_idx = np.argmax(iou, axis=1) # shape: (n_boxes,)
    y_train = np.concatenate([y_train, anchor_idx.reshape(-1, 1)], axis=1)
    # y_train: shape (n_boxes, len([xmin, ymin, xmax, ymax, class, anchor_index]))

    #
    y_outs = []
    for anchor_idxs in anchor_mask:
        # 6 for xmin, ymin, xmax, ymax,objectness score, class
        y_true_out = np.zeros((grid_size, grid_size, anchor_idxs.shape[0], 4+1+14))
        for r in y_train: # iterate through all boxes
            if np.isnan(r[0]) or np.isinf(r[0]):
                continue
            anchor_eq = (anchor_idxs == int(r[-1]))
            if np.any(anchor_eq):
                box = r[:4]
                box_xy = (box[0:2] + box[2:4]) / 2 # box center
                anchor_idx = np.where(anchor_eq)[0][0]
                grid_x, grid_y = (box_xy // (1/grid_size)).astype(int)
                class_probs = np.zeros(14)
                indices = y_train[:,4].astype(int)
                indices = indices[indices!=14]
                class_probs[indices] = 1.
                class_probs = class_probs.tolist()
                y_true_out[grid_x][grid_y][anchor_idx] = [box[0], box[1], box[2], box[3], 1] + class_probs
                
        y_outs.append(y_true_out)
        grid_size *= 2
    
    return y_outs
```

## YOLO Loss

```python
import torch
import torch.nn as nn
```

```python
class YOLOLoss(nn.Module):
    def __init__(self, anchor_boxes, n_classes, iou_threshold=0.5):
        super(YOLOLoss, self).__init__()
        
        self.anchor_boxes = anchor_boxes
        self.n_classes = n_classes
        self.iou_threshold = iou_threshold
        
    def __refine_output__(self, pred, anchors, classes):
        # pred: (batch_size, grid, grid, anchors, (x, y, w, h, obj, ...classes))
        grid_size = pred.shape[1:3]
        box_xy, box_wh, objectness, class_probs = torch.split(pred, (2, 2, 1, classes), dim=-1)
        
        box_xy = torch.sigmoid(box_xy)
        objectness = torch.sigmoid(objectness)
        class_probs = torch.sigmoid(class_probs)
        pred_box = torch.cat((box_xy, box_wh), axis=-1)  # original xywh for loss
        
        grid = torch.meshgrid(torch.arange(0, grid_size[1]), torch.arange(0, grid_size[0]))
        grid = torch.unsqueeze(torch.stack(grid, axis=-1), axis=2)  # shape: [gx, gy, 1, 2]
        
        box_xy = (box_xy + grid) / torch.tensor(grid_size)
        box_wh = torch.exp(box_wh) * anchors
        
        box_x1y1 = box_xy - box_wh / 2
        box_x2y2 = box_xy + box_wh / 2
        bbox = torch.cat([box_x1y1, box_x2y2], axis=-1)

        return bbox, objectness, class_probs, pred_box
        
    def _broadcast_iou(self, box_1, box_2):
        # box_1: (..., (x1, y1, x2, y2))
        # box_2: (N, (x1, y1, x2, y2))
        
        # broadcast boxes
        box_1 = torch.unsqueeze(box_1, -2)
        box_2 = torch.unsqueeze(box_2, 0)
        
        box_1, box_2 = torch.broadcast_tensors(box_1, box_2)

        int_w = torch.minimum(box_1[..., 2], box_2[..., 2]) - torch.maximum(box_1[..., 0], box_2[..., 0])
        int_w = torch.maximum(int_w, torch.FloatTensor([0]).expand_as(int_w))[0]
        print('type(int_w):', type(int_w))
        
        int_h = torch.minimum(box_1[..., 3], box_2[..., 3]) - torch.maximum(box_1[..., 1], box_2[..., 1])
        int_h = torch.maximum(int_h, torch.FloatTensor([0]).expand_as(int_h))[0]
        print('type(int_h):', type(int_h))
        int_area = int_w * int_h
        
        box_1_area = (box_1[..., 2] - box_1[..., 0]) * (box_1[..., 3] - box_1[..., 1])
        box_2_area = (box_2[..., 2] - box_2[..., 0]) * (box_2[..., 3] - box_2[..., 1])
        
        iou = int_area / (box_1_area + box_2_area - int_area)
        print('type(iou):', type(iou))
        
        return iou

        
    def forward(self, y_pred, y_true):
        # 1. transform all pred outputs
        # y_pred: (batch_size, grid, grid, anchors, (x, y, w, h, obj, ...cls))
        pred_box, pred_obj, pred_class, pred_xywh = self.__refine_output__(y_pred, 
                                                                           self.anchor_boxes, 
                                                                           self.n_classes)
        pred_xy = pred_xywh[..., 0:2]
        pred_wh = pred_xywh[..., 2:4]
        
        # 2. transform all true outputs
        # y_true: (batch_size, grid, grid, anchors, (x1, y1, x2, y2, obj, ...cls))
        true_box, true_obj, true_class_idx = torch.split(y_true, (4, 1, self.n_classes), dim=-1)
        true_xy = (true_box[..., 0:2] + true_box[..., 2:4]) / 2
        true_wh = true_box[..., 2:4] - true_box[..., 0:2]

        print('true_box[true_box>0]:', true_box[true_box>0])
        print('true_obj.bool().any():', true_obj.bool().any())
        
        # give higher weights to small boxes
        box_loss_scale = 2 - true_wh[..., 0] * true_wh[..., 1]
        
        # 3. inverting the pred box equations
        grid_size = y_true.shape[1]
        grid = torch.meshgrid(torch.arange(grid_size), torch.arange(grid_size))
        grid = torch.unsqueeze(torch.stack(grid, axis=-1), axis=2)
        true_xy = true_xy * grid_size - grid
        true_wh = torch.log(true_wh / self.anchor_boxes)
        true_wh = torch.where(torch.isnan(true_wh), torch.zeros_like(true_wh), true_wh)
        true_wh = torch.where(torch.isinf(true_wh), torch.zeros_like(true_wh), true_wh)
        
        # 4. calculate all masks
        obj_mask = torch.squeeze(true_obj, -1)
        # ignore false positive when iou is over threshold
        best_iou = self._broadcast_iou(pred_box, true_box[obj_mask.bool()])
        best_iou = torch.max(best_iou, axis=-1)[0]
        ignore_mask = (best_iou < self.iou_threshold)

        # 5. calculate all losses
        xy_loss = obj_mask * box_loss_scale * torch.sum((true_xy - pred_xy)**2, axis=-1)
        wh_loss = obj_mask * box_loss_scale * torch.sum((true_wh - pred_wh)**2, axis=-1)
        obj_loss = nn.BCELoss()(pred_obj, true_obj).item()
        obj_loss = obj_mask * obj_loss + (1 - obj_mask) * ignore_mask * obj_loss
        class_loss = obj_mask * nn.BCELoss()(pred_class, true_class_idx).item()
        
         # 6. sum over (batch, gridx, gridy, anchors) => (batch, 1)
        xy_loss = torch.sum(xy_loss, axis=(1, 2, 3))
        wh_loss = torch.sum(wh_loss, axis=(1, 2, 3))
        obj_loss = torch.sum(obj_loss, axis=(1, 2, 3))
        class_loss = torch.sum(class_loss, axis=(1, 2, 3))

        return xy_loss + wh_loss + obj_loss + class_loss
```
